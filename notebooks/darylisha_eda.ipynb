## What profile characteristics are associated with higher attractiveness or engagement? ##

For our analysis of profile characteristics and their association with higher attractiveness or engagement, we examined the following variables:

- Number of photos per profile
- Preferred gender
- Languages spoken

According to the available dataset, the maximum number of photos on a profile is 21, with the highest average number of photos being 3 for users aged 27. In comparison, the lowest age group, 18, has an average of only 1 profile picture. The preferred gender among users is male, with no preference being a close second. Additionally, the maximum number of spoken languages per profile is only 1, which applies to 3,500 individuals.

We explored the relationships between the following variables:

1) **Age and number of fans**: The pie chart illustrates that 19% of fan counts are from 23-year-olds, followed closely by 15% from 24-year-olds. On the other hand, individuals aged 26-28 have the lowest fan counts, with less than 5%.

2) **Number of photos on the profile and profile visits**: The Pearson coefficient for this relationship is 0.37, indicating a weak positive correlation. This suggests that there is a relationship between the number of photos and profile visits, although it is not very strong.

3) **Number of pictures and number of kisses**: The Pearson coefficient here is 0.09, indicating a very weak positive correlation. This shows that there is minimal relationship between the number of pictures and the number of kisses received on profiles.

4) **Number of languages spoken and gifts**: The Pearson coefficient is 0.03, suggesting no significant relationship between the number of languages spoken and the number of gifts received. Therefore, there is little to no correlation between these two variables.

5) **Language spoken and number of fans**: The Pearson coefficient is -0.70, indicating a strong negative correlation between the number of fans on a user’s profile and the language spoken. 

We found varying degrees of correlation among the variables we examined, shedding light on how profile characteristics may influence engagement and attractiveness.

In summary, there is a weak positive correlation between the number of photos on a user’s profile and the number of visits and kisses each profile receives. Generally, having more profile pictures or a more detailed profile can lead to an increase in the number of users viewing your profile. However, it's important to note that our dataset did not include information about the dating status or outcomes of the profiles. As a result, we could not determine whether certain characteristics had an impact on dating success.



## Final Thoughts ##

After completing our analysis, we discovered a lack of variety in the data for most profiles on Lovoo. The dataset's author noted in the methodology section that they could only browse user profiles recommended by Lovoo's algorithm. Even after creating two profiles, the same user profiles continually appeared. Consequently, only female profiles were included in the CSV file.

Additionally, since the app is region-based, there is a bias in the representation of male users and other gender identities compared to female users. The data collected cannot be used to determine a causal relationship between dating apps and their impact on dating profiles or outcomes. Overall, a more comprehensive dataset that includes additional categories such as education, hobbies, and interests would provide better insights into what makes a profile more attractive to other users.
